```python
# Total Centuries
centuries_count=centuries['Runs'].count()
centuries_count
```

```python
# Most Centuries
most_centuries_count=centuries['Player'].value_counts()
most_centuries_count.head()
```

```python
# Most Centuries Bar Plot
most_centuries = centuries['Player'].value_counts().head(10)
most_centuries_name = most_centuries.index
last_names = [name.split()[-1] for name in most_centuries_name]
most_centuries_count = most_centuries.values

x = np.array(last_names)
y = np.array(most_centuries_count)

plt.figure(figsize=(10,6))
plt.title("Most Centuries")
plt.bar(x,y,color='red')
plt.show()
```

```python
# Most Centuries in Winning Cause
most_centuries_count=centuries[(centuries['Result']=='Won')]['Player'].value_counts()
most_centuries_count.head(10)
```

```python
# Most Centuries in Winning Cause in Bar Plot
most_centuries = centuries[(centuries['Result']=='Won')]['Player'].value_counts().head(10)
most_centuries_name = most_centuries.index
last_names = [name.split()[-1] for name in most_centuries_name]
most_centuries_count = most_centuries.values

x = np.array(last_names)
y = np.array(most_centuries_count)

plt.figure(figsize=(10,6))
plt.title("Most Centuries in Winning Cause")
plt.bar(x,y,color='green')
plt.show()
```
